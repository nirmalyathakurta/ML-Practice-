```python
import pandas as pd
from apyori import apriori
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
```

```python
df.head(10)
```

```python
df.describe()
df.isnull().sum()
```

```python
transactions = []
for i in range(len(df)):
    transactions.append([str(df.values[i, j]) for j in range(df.shape[1]) if str(df.values[i, j]) != 'nan'])
```

```python
rules = apriori(transactions, min_support=0.003, min_confidence=0.2, min_lift=3, min_length=2)
```

```python
results = list(rules)
print("Number of association rules generated:", len(results))
for rule in results:
    print(rule)
```

```python
rules_high_confidence = apriori(transactions, min_support=0.003, min_confidence=0.5, min_lift=3, min_length=2)
results_high_confidence = list(rules_high_confidence)
print("Number of association rules generated with high confidence:", len(results_high_confidence))
for rule in results_high_confidence:
    print(rule)
```
